```python
cgdf["Year"] = cgdf["Year"].astype(int)
```

# Data Visualization

```python
plt.figure(figsize = (12, 3))
sns.barplot(x = "Genre", y = "Name", data = cgdf, palette = "rainbow")
```

```python
corr_matrix = cgdf.corr(numeric_only = True)
plt.figure(figsize = (8, 5))
sns.heatmap(corr_matrix, annot = True)
```
